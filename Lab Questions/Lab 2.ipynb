```python
import time
import matplotlib.pyplot as plt


def recursive_binary_search(arr, low, high, target):
    """
    A recursive implementation of binary search.
    """
    if high >= low:
        mid = low + (high - low) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return recursive_binary_search(arr, low, mid - 1, target)
        else:
            return recursive_binary_search(arr, mid + 1, high, target)
    else:
        return -1

# --- Experiment Setup ---
# Different values of n (number of elements)
n_values = [1000, 10000, 100000, 1000000, 5000000, 10000000]
times = [] # To store the execution time for each n

# The element to search for (worst-case: element not in the list)
target_to_find = -1

print("Running experiment...")
for n in n_values:
    # 1. Create a sorted list of size n
    sorted_list = list(range(n))

    # 2. Measure the time taken
    start_time = time.perf_counter()
    recursive_binary_search(sorted_list, 0, len(sorted_list)-1, target_to_find)
    end_time = time.perf_counter()

    # Calculate and store the duration
    duration = end_time - start_time
    times.append(duration)
    print(f"n = {n:<10} | Time taken: {duration:.8f} seconds")

print("\nExperiment finished.")

# --- Plotting the Results ---
plt.figure(figsize=(10, 6))
plt.plot(n_values, times, marker='o', linestyle='-', color='b')
plt.xlabel("Number of Elements (n)")
plt.ylabel("Time Taken (seconds)")
plt.title("Recursive Binary Search Performance")
plt.grid(True)
plt.xscale('log') # Use a logarithmic scale for the x-axis
plt.yscale('log') # Use a logarithmic scale for the y-axis
plt.show()
```
