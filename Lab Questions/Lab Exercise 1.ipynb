```python
import random
import time
import matplotlib.pyplot as plt

# Linear search
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Perform timing for a given position
def measure_search_times(position, sequence_lengths, trials):
    times = []

    for n in sequence_lengths:
        total_time = 0

        for _ in range(trials):
            arr = random.sample(range(1, n * 10), n)

            if position == 'Start':
                target = arr[0]
            elif position == 'Middle':
                target = arr[n // 2]
            elif position == 'End':
                target = arr[-1]
            else:
                raise ValueError("Invalid position")

            start = time.perf_counter()
            linear_search(arr, target)
            end = time.perf_counter()

            total_time += (end - start)

        average_time = (total_time / trials) * 1000  # ms
        times.append(average_time)

    return times

# Plot for one position
def plot_search_case(sequence_lengths, times, position):
    plt.plot(sequence_lengths, times, marker='o')
    plt.xlabel("Input Size (n)")
    plt.ylabel("Average Time (ms)")
    plt.title(f"Search Time  {position}")
    plt.grid(True)
    plt.show()

# Main function
def run_search_analysis():
    sequence_lengths = [5000, 10000, 15000, 20000]
    trials = 100

    for position in ['Start', 'Middle', 'End']:
        times = measure_search_times(position, sequence_lengths, trials)
        plot_search_case(sequence_lengths, times, position)

# Execute
run_search_analysis()
```
