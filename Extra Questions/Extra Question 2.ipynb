```python
import time
import random
import matplotlib.pyplot as plt

# Linear search that returns index and comparison count
def linear_search(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return comparisons
    return  comparisons

# Input sizes to test
input_sizes = [100, 500, 1000, 5000, 10000]

# Prepare data structures
best_case_times = []
average_case_times = []
worst_case_times = []

best_case_comparisons = []
average_case_comparisons = []
worst_case_comparisons = []

# Run for each input size
for n in input_sizes:
    values = random.sample(range(1, n * 10), n)  # Random list

    # Define targets
    targets = [
        values[0],            # Best Case
        values[n // 2],       # Average Case
        -1                    # Worst Case (not in list)
    ]

    # Run all 3 cases for current n
    for i, case in enumerate(["Best", "Average", "Worst"]):
        start = time.perf_counter()
        comparisons = linear_search(values, targets[i])
        end = time.perf_counter()
        elapsed_time = (end - start) * 1000  # ms

        # Append to appropriate list
        if case == "Best":
            best_case_times.append(elapsed_time)
            best_case_comparisons.append(comparisons)
        elif case == "Average":
            average_case_times.append(elapsed_time)
            average_case_comparisons.append(comparisons)
        else:  # Worst
            worst_case_times.append(elapsed_time)
            worst_case_comparisons.append(comparisons)

# Print Table
print("\n=== Linear Search Results ===")
print(f"{'Input Size':<12}{'Best Time':<12}{'Avg Time':<12}{'Worst Time':<12}"
      f"{'Best Comp':<12}{'Avg Comp':<12}{'Worst Comp'}")

for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:<12}"
          f"{best_case_times[i]:<12.5f}"
          f"{average_case_times[i]:<12.5f}"
          f"{worst_case_times[i]:<12.5f}"
          f"{best_case_comparisons[i]:<12}"
          f"{average_case_comparisons[i]:<12}"
          f"{worst_case_comparisons[i]}")

# Plot Time Graph
plt.plot(input_sizes, best_case_times, marker='o', label='Best Case Time', color='green')
plt.plot(input_sizes, average_case_times, marker='s', label='Average Case Time', color='orange')
plt.plot(input_sizes, worst_case_times, marker='^', label='Worst Case Time', color='red')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (ms)")
plt.title("Search Time vs Input Size")
plt.legend()
plt.grid(True)
plt.show()

# Plot Comparison Count Graph
plt.plot(input_sizes, best_case_comparisons, marker='o', label='Best Case Comparisons', color='green')
plt.plot(input_sizes, average_case_comparisons, marker='s', label='Average Case Comparisons', color='orange')
plt.plot(input_sizes, worst_case_comparisons, marker='^', label='Worst Case Comparisons', color='red')
plt.xlabel("Input Size (n)")
plt.ylabel("Number of Comparisons")
plt.title("Comparison Count vs Input Size")
plt.legend()
plt.grid(True)
plt.show()

print("\n2. Growth of Comparisons:")
print("-> Best Case: Always 1 comparison (constant).")
print("-> Average Case: Grows approximately with n/2.")
print("-> Worst Case: Grows linearly with n.")

print("\n3. Comparison with Theoretical Expectations:")
print("-> Best Case     ≈ 1 comparison    → O(1)")
print("-> Average Case  ≈ n/2 comparisons → O(n)")
print("-> Worst Case    ≈ n comparisons   → O(n)")
```
