```python
import time
import random
import matplotlib.pyplot as plt

# Linear Search Function
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Setup
cases = ["Best Case", "Average Case", "Worst Case"]
times = []
n = 10000
values = random.sample(range(1, n * 2), n)  # Unique random numbers

# Targets for Best, Average, Worst
targets = [
    values[0],                  # Best Case: element at beginning
    values[len(values)//2],     # Average Case: element in middle
    -1                          # Worst Case: element not in list
]

# Measure time for each case
for target in targets:
    start = time.perf_counter()
    linear_search(values, target)
    end = time.perf_counter()
    
    search_time = (end - start) * 1000  # in milliseconds
    times.append(search_time)

# Plotting the results
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.xlabel("Cases")
plt.ylabel("Search Time (ms)")
plt.title("Time Complexity Analysis – Linear Search")
plt.tight_layout()
plt.show()

# Printing actual time results
print("=== Execution Time Results ===")
print(f"Best Case Time   : {times[0]:.5f} ms")
print(f"Average Case Time: {times[1]:.5f} ms")
print(f"Worst Case Time  : {times[2]:.5f} ms")

# Explanations
print("\n=== Time Complexity Analysis ===")
print("Best Case (O(1)): Element is found at the first index. Only one comparison is made.")
print("Average Case (O(n/2)): Element is somewhere in the middle. On average, n/2 comparisons.")
print("Worst Case (O(n)): Element is not found. All n elements are compared.")

# Observation based on input size
print("\n=== Observation on Input Size (n = 10000) ===")
print("As n increases, the search time increases linearly in the worst case.")
print("This is expected since linear search checks elements one-by-one.")
```
